```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier, VotingClassifier
from sklearn.tree import DecisionTreeClassifier
import re
import string
```

```python
df_fake = pd.read_csv("Fake.csv")
df_true = pd.read_csv("True.csv")
```

```python
df_fake=df_fake.head(1000)
df_fake
```

```python
df_true=df_true.head(1000)
df_true
```

```python
df_fake["class"] = 0
df_true["class"] = 1
```

```python
df_fake.shape, df_true.shape
```

```python
df_merge = pd.concat([df_fake, df_true], axis =0 )
df_merge.head(10)
```

```python
df = df_merge.drop(["title", "subject","date"], axis = 1)
```

```python
df = df.sample(frac = 1)
```

```python
df.head()
```

```python
df.reset_index(inplace = True)
df.drop(["index"], axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.head()
```

```python
df.to_csv('finaldataset.csv',index=False)
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) 
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)    
    return text
```

```python
df["text"] = df["text"].apply(wordopt)
```

```python
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
import pickle
with open('tfidf_vectorizer.pkl', 'wb') as f:
    pickle.dump(vectorization, f)
```

```python
from sklearn.metrics import accuracy_score, classification_report
knn = KNeighborsClassifier(n_neighbors=5)
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
ensemble = VotingClassifier(estimators=[
    ('knn', knn),
    ('gb', gb),
    ('dt', dt)
], voting='hard')

ensemble.fit(xv_train, y_train)
ensemble_pred = ensemble.predict(xv_test)


print("\nEnsemble Accuracy:", accuracy_score(y_test, ensemble_pred))
```

```python
with open('voting_classifier.pkl', 'wb') as f:
    pickle.dump(ensemble, f)
```

```python
with open('tfidf_vectorizer.pkl', 'rb') as f:
    loaded_vectorizer = pickle.load(f)
```
